```python
def compute_cum_goal_band(df, window, q_low, q_high):
    df = df.sort_values(["run_id", "step"])
    df["rolling_mean"] = df.groupby("run_id")["value"].transform(
        lambda x: x.rolling(window=window, min_periods=1).mean()
    )
    df["cummean"] = (
        df.groupby("run_id")["rolling_mean"]
        .expanding()
        .mean()
        .reset_index(level=0, drop=True)
    )
    medians = df.groupby(["step"])["cummean"].median()
    quantile_low = df.groupby(["step"])["cummean"].quantile(q_low)
    quantile_high = df.groupby(["step"])["cummean"].quantile(q_high)
    return medians, quantile_low, quantile_high


uw_td3_median, uw_td3_low, uw_td3_high = compute_cum_goal_band(
    td3_goal_reaching.copy(), window, q_low, q_high
)
uw_rrl_median, uw_rrl_low, uw_rrl_high = compute_cum_goal_band(
    rrl_goal_reaching.copy(), window, q_low, q_high
)
uw_calf_median, uw_calf_low, uw_calf_high = compute_cum_goal_band(
    calf_goal_reaching.copy(), window, q_low, q_high
)
```

```python
def baseline_policy_band(
    csv_path,
    key="episode_stats/n_safe_actions",
    grid_step=1000,
    grid_start=50_000,
    rolling_window=5,
):
    df = pd.read_csv(csv_path)
    df = df[df["key"] == key].copy()
    df["value"] = pd.to_numeric(df["value"], errors="coerce")
    df = df.dropna(subset=["value"]).sort_values(["run_id", "step"])

    grid_begin = max(grid_start, int(df["step"].min()))
    grid = np.arange(grid_begin, df["step"].max() + 1, grid_step)
    grid_df = pd.DataFrame({"step": grid}).sort_values("step")

    def ffill_to_grid(run_id, g):
        g = g.sort_values("step")[["step", "value"]]
        out = pd.merge_asof(grid_df, g, on="step", direction="backward")
        out["run_id"] = run_id
        return out

    interp = pd.concat(
        [ffill_to_grid(run_id, g) for run_id, g in df.groupby("run_id")],
        ignore_index=True,
    )

    means = (
        interp.groupby("step")["value"]
        .mean()
        .rolling(window=rolling_window, min_periods=1)
        .mean()
    )
    stds = (
        interp.groupby("step")["value"]
        .std()
        .rolling(window=rolling_window, min_periods=1)
        .mean()
    )
    return means * 100, stds * 100


uw_means, uw_stds = baseline_policy_band("expdata/baselinepolicy_fraction.csv")
robot_means, robot_stds = baseline_policy_band(
    "expdata/baselinepolicy_fraction_robot.csv"
)

fig, axes = plt.subplots(1, 2, figsize=(6, 3.0), sharey=True)
ax_left, ax_right = axes

# Underwater drone (left)
ax_left.plot(uw_means)
ax_left.fill_between(uw_means.index, uw_means - uw_stds, uw_means + uw_stds, alpha=0.2)
ax_left.axvline(
    x=0.9 * 3_000_000, color="black", linestyle="--", label="Baseline disabled"
)
ax_left.set_xlim(50_000, 3_000_000)
ax_left.set_title("Underwater drone")
ax_left.set_xlabel("Step")
ax_left.grid()

# Kinematic robot (right)
ax_right.plot(robot_means)
ax_right.fill_between(
    robot_means.index,
    robot_means - robot_stds,
    robot_means + robot_stds,
    alpha=0.2,
)
ax_right.axvline(x=0.9 * 3_000_000, color="black", linestyle="--", label="_nolegend_")
ax_right.set_xlim(50_000, 3_000_000)
ax_right.set_title("Kinematic robot")
ax_right.set_xlabel("Step")
ax_right.grid()

fig.supylabel(r"Fraction of baseline policy\\calls to total episode steps (in \%)")
handles, labels = ax_left.get_legend_handles_labels()
fig.legend(handles, labels, loc="upper center", ncol=1, bbox_to_anchor=(0.5, 1.03))
try:
    fig.tight_layout(rect=[0, 0, 1, 0.95])
except RuntimeError:
    # Fallback when decorations (suplabel/legend) make tight_layout fail
    fig.subplots_adjust(top=0.82, bottom=0.18, wspace=0.25)
fig.savefig("../gfx/baseline_policy_calls_percent.pdf")
```

```python
from pathlib import Path

calfq_data = pd.read_csv("expdata/cleared/robot/calf.csv")
residual_td3_data = pd.read_csv("expdata/cleared/robot/residual_td3.csv")
td3_data = pd.read_csv("expdata/cleared/robot/td3.csv")

episodic_return_calfq = select_metrics(calfq_data, "charts/episodic_return")
episodic_return_td3 = select_metrics(td3_data, "charts/episodic_return")
episodic_return_residual_td3 = select_metrics(
    residual_td3_data, "charts/episodic_return"
)

goal_reaching_calfq = select_metrics(calfq_data, "episode_stats/goal_reached")
goal_reaching_td3 = select_metrics(td3_data, "episode_stats/goal_reached")
goal_reaching_residual_td3 = select_metrics(
    residual_td3_data, "episode_stats/goal_reached"
)


def rolling_stats(df, rolling_window=100, grid_step=5000):
    # 1) rolling 100 внутри каждого run_id по step
    df = df.sort_values(["run_id", "step"])
    df["roll100"] = df.groupby("run_id")["value"].transform(
        lambda s: s.rolling(rolling_window, min_periods=1).mean()
    )

    # 2) ffill на сетку шагов каждые 1000
    grid = np.arange(df["step"].min(), df["step"].max() + 1, grid_step)
    grid_df = pd.DataFrame({"step": grid}).sort_values("step")

    def ffill_to_grid(run_id, g):
        g = g.sort_values("step")[["step", "roll100"]]
        out = pd.merge_asof(grid_df, g, on="step", direction="backward")
        out["run_id"] = run_id
        return out

    interp = pd.concat(
        [ffill_to_grid(run_id, g) for run_id, g in df.groupby("run_id")],
        ignore_index=True,
    )

    # 3) mean по шагам
    mean_by_step = interp.groupby("step")["roll100"].mean()

    # 4) std по шагам
    std_by_step = interp.groupby("step")["roll100"].std()
    return mean_by_step, std_by_step
```

```python
print("ROBOT GOAL REACHING RATE AFTER 2.7M STEPS (closer to 1 is better)")


def print_metric_rate(df, label):
    df = df.copy()
    df["value"] = pd.to_numeric(df["value"], errors="coerce")
    df = df.dropna(subset=["value"])
    df = df[df["step"] >= 2_700_000]
    if df.empty:
        print(label)
        print("no data after 2.7M steps")
        return
    per_run = df.groupby("run_id")["value"].mean()
    print(label)
    print(per_run.mean(), "+-", per_run.std())


print_metric_rate(goal_reaching_calfq, "CALFQ")
print_metric_rate(goal_reaching_td3, "TD3")
print_metric_rate(goal_reaching_residual_td3, "TD3 RESIDUAL")

print("ROBOT TARGETS CAPTURED RATE AFTER 2.7M STEPS (closer to 1 is better)")

targets_captured_calfq = select_metrics(calfq_data, "episode_stats/targets_captured")
targets_captured_td3 = select_metrics(td3_data, "episode_stats/targets_captured")
targets_captured_residual_td3 = select_metrics(
    residual_td3_data, "episode_stats/targets_captured"
)

print_metric_rate(targets_captured_calfq, "CALFQ")
print_metric_rate(targets_captured_td3, "TD3")
print_metric_rate(targets_captured_residual_td3, "TD3 RESIDUAL")

print("ROBOT COMBINED RATE AFTER 2.7M STEPS (closer to 1 is better)")


def print_combined_rate(goal_df, target_df, label):
    goal_df = goal_df.copy()
    target_df = target_df.copy()
    goal_df["value"] = pd.to_numeric(goal_df["value"], errors="coerce")
    target_df["value"] = pd.to_numeric(target_df["value"], errors="coerce")
    goal_df = goal_df.dropna(subset=["value"])
    target_df = target_df.dropna(subset=["value"])
    goal_df = goal_df[goal_df["step"] >= 2_700_000]
    target_df = target_df[target_df["step"] >= 2_700_000]
    if goal_df.empty or target_df.empty:
        print(label)
        print("no data after 2.7M steps")
        return
    goal_per_run = goal_df.groupby("run_id")["value"].mean()
    target_per_run = target_df.groupby("run_id")["value"].mean()
    aligned = (goal_per_run + target_per_run) / 2
    aligned = aligned.dropna()
    if aligned.empty:
        print(label)
        print("no aligned runs after 2.7M steps")
        return
    print(label)
    print(aligned.mean(), "+-", aligned.std())


print_combined_rate(goal_reaching_calfq, targets_captured_calfq, "CALFQ")
print_combined_rate(goal_reaching_td3, targets_captured_td3, "TD3")
print_combined_rate(
    goal_reaching_residual_td3, targets_captured_residual_td3, "TD3 RESIDUAL"
)
```
